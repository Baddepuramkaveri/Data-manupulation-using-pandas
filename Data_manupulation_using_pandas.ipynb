Handling missing values

```python
#Detecting missing values
missing_values = data.isnull()
missing_values
```

```python
#Remove rows with missing values
data_drop = data.dropna()
data_drop
```

```python
#Remove columns with missing values
data_drop = data.dropna(axis = 1)
data_drop
```

```python
#filling new value in place of null values
data_fill = data.fillna(value = 0)
data_fill
```

```python
#handling missing value conditionally
data_mean = data['ChipRate'].mean()
data_mean
```

```python
data_fill = data.fillna(data_mean)
data_fill
```
